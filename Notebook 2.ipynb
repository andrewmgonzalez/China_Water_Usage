```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_og = pd.read_csv('Clean_Data_Final.csv')
#df_og.columns
```

```python
combo = ['Year',
         '1.1. Total production capacity (10,000 cubic metres/day)',
         '2.1. Total water supply volume / year (10,000 cubic metres/year)',
         '5.2. Total city population (10,000s)',
         '5.3. Total city population with water access (10,000s)',
         '8.1. Total pipe length (km)',
         '17.2. Electricity use for water production \n(1,000 kWh/1,000 cubic metre)',]
df=df_og[combo].copy()
df.dropna(axis=0,how='any',inplace=True)
df.reset_index(drop=True,inplace=True)
df.columns=['Year','Production Capacity','Water Supply per Year','Population','Population w Water Access','Pipe Length','Electricity for Water Production']
df

sns.pairplot(data=df,hue='Year', kind='scatter',diag_kind='hist')
plt.show()
```

```python
combo = ['Province',
         '1.1. Total production capacity (10,000 cubic metres/day)',
         '2.1. Total water supply volume / year (10,000 cubic metres/year)',
         '5.2. Total city population (10,000s)',
         '5.3. Total city population with water access (10,000s)',
         '8.1. Total pipe length (km)',
         '17.2. Electricity use for water production \n(1,000 kWh/1,000 cubic metre)',]
df=df_og[combo].copy()
df.dropna(axis=0,how='any',inplace=True)
df.reset_index(drop=True,inplace=True)
df.columns=['Province','Production Capacity','Water Supply per Year','Population','Population w Water Access','Pipe Length','Electricity for Water Production']
df

sns.pairplot(data=df,hue='Province', kind='scatter',diag_kind='hist')
plt.show()
```
